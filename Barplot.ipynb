```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
sns.barplot(x='who',y='fare',data=df)
```

```python
sns.barplot(x='who',y='fare',data=df)
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=df,color='magenta')
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=df,color='magenta',hue='sex')
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=df,color='green',hue='sex')
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=df,color='green',hue='sex',palette='spring')
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=df,color='green',hue='sex',palette='spring',saturation=0.2)
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=df,color='green',hue='sex',palette='spring',saturation=0.2,ci=False)
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=df,color='green',hue='sex',palette='spring',saturation=0.2,order=['child','woman','men'])
plt.show()
```
