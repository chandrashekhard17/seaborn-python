```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.violinplot(x=df['total_bill'])
plt.show()
```

```python
sns.violinplot(y='tip',data=df)
```

```python
sns.violinplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
sns.violinplot(x='sex',y='tip',data=df)
plt.show()
```

```python
sns.violinplot(x='sex',y='tip',data=df,hue='day')
plt.show()
```

```python
sns.violinplot(x='sex',y='tip',data=df,hue='day',color='hotpink')
plt.show()
```

```python
sns.violinplot(x='sex',y='tip',data=df,hue='day',palette='viridis')
plt.show()
```

```python
sns.violinplot(x='sex',y='tip',data=df,palette='viridis')
plt.show()
```

```python
sns.violinplot(y='sex',x='tip',data=df,palette='viridis')
plt.show()
```

```python
sns.violinplot(x='sex',y='tip',data=df,hue='day',palette='viridis',saturation=1)
plt.show()
```

```python
sns.violinplot(x='sex',y='tip',data=df,hue='day',palette='viridis',saturation=1,linewidth=3)
plt.show()
```
