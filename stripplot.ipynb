```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.stripplot(x=df['tip'],data=df)
```

```python
sns.stripplot(x='day',y='tip',data=df)
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=df,hue='sex')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=df,hue='sex',jitter=0.8)
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=df,hue='sex',jitter=0.3)
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=df,hue='sex',jitter=0.8,palette='GnBu')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=df,hue='sex',jitter=0.8,marker='*')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=df,hue='sex',jitter=0.8,marker='*',size=3)
plt.show()
```
