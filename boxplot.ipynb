```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.boxenplot(x='tip',data=df)
```

```python
sns.boxplot(x='tip',data=df)
plt.show()
```

```python
sns.boxplot(x='sex',y='tip',data=df)
plt.show()
```

```python
sns.boxplot(x='sex',y='tip',data=df,hue='smoker')
plt.show()
```

```python
sns.boxplot(x='sex',y='tip',data=df,hue='smoker',palette='viridis')
plt.show()
```

```python
sns.boxplot(x='sex',y='tip',data=df,hue='smoker',color='hotpink')
plt.show()
```

```python
sns.boxplot(x='sex',y='tip',data=df,hue='smoker',color='hotpink',linewidth=3)
plt.show()
```

```python
sns.boxplot(x='tip',data=df,orient='h')
plt.show()
```

```python
sns.boxplot(x='tip',data=df,orient='v')
plt.show()
```
