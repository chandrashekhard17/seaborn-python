```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.histplot(x='total_bill',data=df,discrete='false')
plt.show()
```

```python
sns.histplot(x='total_bill',data=df)
plt.show()
```

```python
sns.histplot(x='total_bill',data=df,discrete='false',hue='sex')
plt.show()
```

```python
sns.histplot(x='total_bill',data=df,discrete='false',hue='sex',palette='spring')
plt.show()
```

```python
sns.histplot(x='total_bill',data=df,discrete='false',hue='sex',kde='True')
plt.show()
#kde--kernel density estimation
```

```python
sns.histplot(x='total_bill',data=df,discrete='false',hue='sex',edgecolor='red')
plt.show()
```

```python
sns.histplot(x='total_bill',data=df,discrete='false',hue='sex',kde='True',edgecolor='violet')
plt.show()
#kde--kernel density estimation
```

```python
sns.histplot(x='total_bill',data=df,discrete='false',hue='sex',kde='True',edgecolor='violet',linestyle='--')
plt.show()
#kde--kernel density estimation
```
