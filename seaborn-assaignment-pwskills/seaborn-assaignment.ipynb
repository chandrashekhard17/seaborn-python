```python
#1...
"""Scatter Plot (sns.scatterplot)

Use: A scatter plot displays the relationship between two continuous variables. Each point represents an observation in the dataset.
Example: Analyzing the relationship between height and weight in a population.
Histogram (sns.histplot)

Use: A histogram shows the distribution of a single continuous variable by dividing the data into bins and counting the number of observations in each bin.
Example: Visualizing the distribution of ages in a survey dataset.
Box Plot (sns.boxplot)

Use: A box plot (or box-and-whisker plot) summarizes the distribution of a dataset by showing the median, quartiles, and potential outliers. It is useful for comparing distributions between different groups.
Example: Comparing the test scores of students from different schools.
Heatmap (sns.heatmap)

Use: A heatmap is a graphical representation of data where individual values are represented as colors. It is often used to display the correlation matrix of a dataset.
Example: Visualizing the correlation between different features in a dataset.
Line Plot (sns.lineplot)

Use: A line plot displays data points connected by a continuous line and is useful for visualizing trends over time or another continuous variable.
Example: Plotting the trend of stock prices over a year."""
```

```python
#2...
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)

# Add title and labels
plt.title("FMRI Signal over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

```python
#3...
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_df = sns.load_dataset("titanic")

# Create the first box plot: pclass vs. age
plt.figure(figsize=(10, 6))
sns.boxplot(x="pclass", y="age", data=titanic_df)
plt.title("Box Plot: pclass vs. age")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()

# Create the second box plot: pclass vs. fare
plt.figure(figsize=(10, 6))
sns.boxplot(x="pclass", y="fare", data=titanic_df)
plt.title("Box Plot: pclass vs. fare")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
#4...
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_df = sns.load_dataset("diamonds")

# Create a histogram for the 'price' column with hue='cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_df, x="price", hue="cut", kde=True, bins=30)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend(title="Cut")
plt.show()
```

```python
#5...
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris_df = sns.load_dataset("iris")

# Create a pair plot with hue='species'
sns.pairplot(data=iris_df, hue="species")
plt.title("Pair Plot of Iris Dataset by Species")
plt.show()
```

```python
#6...
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_df = sns.load_dataset("flights")

# Pivot the data to create a matrix of passenger counts by month and year
flights_pivot = flights_df.pivot(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", cbar_kws={"label": "Passenger Count"})
plt.title("Heatmap of Passenger Counts by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
