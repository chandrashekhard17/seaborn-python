```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.catplot(x='sex',y='tip',data=df)
plt.show()
```

```python
sns.catplot(y='sex',x='tip',data=df)
plt.show()
```

```python
sns.catplot(x='sex',y='tip',data=df,kind='bar')
plt.show()
```

```python
sns.catplot(x='sex',y='tip',data=df,kind='violin')
plt.show()
```

```python
sns.catplot(x='sex',y='tip',data=df,kind='strip')
plt.show()
```

```python
sns.catplot(x='sex',y='tip',data=df,kind='box')
plt.show()
```

```python
sns.catplot(x='sex',y='tip',data=df,kind='bar')
plt.show()
```

```python
sns.catplot(x='sex',y='tip',data=df,kind='bar',hue='smoker')
plt.show()
```

```python
sns.catplot(x='sex',y='tip',data=df,kind='bar',hue='smoker',palette='viridis')
plt.show()
```

```python
sns.catplot(x='sex',y='tip',data=df,kind='bar',hue='smoker',palette='viridis',col='time')
plt.show()
```
