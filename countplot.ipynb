```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.countplot(x='tip',data=df)
plt.show()
```

```python
sns.countplot(x='sex',data=df)
plt.show()
```

```python
df.head()
```

```python
sns.countplot(y='total_bill',data=df)
plt.show()
```

```python
sns.countplot(x='tip',data=df,hue='smoker')
plt.show()
```

```python
sns.countplot(x='tip',data=df,hue='smoker',palette='viridis')
plt.show()
```

```python
sns.countplot(x='tip',data=df,hue='smoker',color='hotpink')
plt.show()
```

```python
sns.countplot(x='tip',data=df,hue='smoker',color='hotpink',saturation=1)
plt.show()
```

```python
sns.countplot(x='tip',data=df,facecolor=[0,0,0,0],linewidth=4,edgecolor=sns.color_palette('GnBu'))
plt.show()
```
