```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
print(data)
```

```python
sns.heatmap(data=data)
```

```python
sns.heatmap(data=data,cmap='spring')
plt.show()
```

```python
sns.heatmap(data=data,cmap='GnBu')
plt.show()
```

```python
sns.heatmap(data=data,cmap='GnBu',alpha=0.5)
plt.show()
```

```python
sns.heatmap(data=data,cmap='GnBu',alpha=0.5,linewidth=2,linecolor='yellow')
plt.show()
```

```python
sns.heatmap(data=data,cmap='GnBu',alpha=0.5,linewidth=2,linecolor='yellow',annot=True)
plt.show()
```

```python
sns.heatmap(data=data,cmap='GnBu',alpha=0.5,linewidth=2,linecolor='yellow',annot=True,cbar=False)
plt.show()
```

```python
sns.heatmap(data=data,cmap='GnBu',alpha=0.5,linewidth=2,linecolor='yellow',annot=True,cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```
